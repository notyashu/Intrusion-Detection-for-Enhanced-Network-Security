```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data for top 20 features
data = {
    'Rank': list(range(1, 21)),
    'Feature': [
        'Bwd Packet Length Std', 'Flow Bytes/s', 'Total Length of Fwd Packets',
        'Fwd Packet Length Std', 'Flow IAT Std', 'Flow IAT Min', 'Fwd IAT Total',
        'Flow Duration', 'Bwd Packet Length Max', 'Flow IAT Max', 'Flow IAT Mean',
        'Total Length of Bwd Packets', 'Fwd Packet Length Min', 'Bwd Packet Length Mean',
        'Flow Packets/s', 'Fwd Packet Length Mean', 'Total Backward Packets',
        'Total Fwd Packets', 'Fwd Packet Length Max', 'Bwd Packet Length Min'
    ],
    'Normalized %': [
        38.97, 28.25, 16.18, 10.10, 1.56, 1.10, 0.81, 0.66, 0.63, 0.57,
        0.52, 0.21, 0.11, 0.09, 0.09, 0.08, 0.03, 0.02, 0.02, 0.01
    ]
}
```

```python
# Create DataFrame and compute cumulative percentage
df = pd.DataFrame(data)
df['Cumulative %'] = df['Normalized %'].cumsum()

# Display the table
print("\nTop-20 Features: Normalized and Cumulative Importances\n")
print(df.to_string(index=False))

# Plot 1: Bar chart of Normalized %
plt.figure()
plt.bar(df['Rank'], df['Normalized %'])
plt.title('Normalized Importance (%) of Top 20 Features')
plt.xlabel('Rank')
plt.ylabel('Normalized %')
plt.xticks(df['Rank'])
plt.tight_layout()
plt.show()

# Plot 2: Line chart of Cumulative %
plt.figure()
plt.plot(df['Rank'], df['Cumulative %'], marker='o')
plt.title('Cumulative Importance (%) of Top 20 Features')
plt.xlabel('Rank')
plt.ylabel('Cumulative %')
plt.xticks(df['Rank'])
plt.tight_layout()
plt.show()
```
